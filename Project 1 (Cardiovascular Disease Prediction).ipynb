```python
# 2.Write a python program to find out and display horizontal bar chart the alcohol consumption details in the year '1986' where WHO region is 'Western Pacific' and country is 'VietNam' from thedataset
df[df ['Year']==1987]
```

```python
res = df[df['Year']==1987].where((df['WHO region']=='Western Pacific') &
                           (df['Country']=='Viet Nam'))
res= res.dropna()
res
```

```python
import matplotlib.pyplot as plt
plt.bar('Beverage types', 'Display Value', data=res, color='orange')
plt.xlabel('alcohol consumption')
plt.ylabel('western pacific - Viet Nam')
plt.show()
```
